```python
TP
```

```python
FP = len(d1[(d1["Actual"]==0) & (d1["Predicted"]==1)])
```

```python
FP
```

```python
FN = len(d1[(d1["Actual"]==1) & (d1["Predicted"]==0)])
```

```python
FN
```

```python
TN = len(d1[(d1["Actual"]==0) & (d1["Predicted"]==0)])
```

```python
TN
```

```python
accuracy_score = (TP+TN)/(TP+TN+FN+FP)
```

```python
accuracy_score
```

```python
precision=TP/(TP+TN)
```

```python
precision
```

```python
recall=TP/(TP+FN)
```
