```python
final=pd.DataFrame()

for j in range(1,50): ## we have to do the 500
    response=requests.get('https://api.themoviedb.org/3/movie/popular?api_key=1e0abc1f4bb5f5024da033cec3b8a2e2&language=en-US&page={}'.format(j)).json()
    
    Popularity=[]
    Vote_count=[]
    Original_language=[]
    Title=[]
    Vote_average=[]
    Overview=[]
    Release_date=[]

    ## you need to understand each and every thing deeply man here go back the same process::
    for i in response["results"]:
        Popularity.append(i["popularity"])
        Vote_count.append(i["vote_count"])
        Original_language.append(i["original_language"])
        Title.append(i["title"])
        Vote_average.append(i["vote_average"])
        Overview.append(i["overview"])
        Release_date.append(i["release_date"])

    d={"Title":Title ,"Overview":Overview , "Release_date":Release_date,"Popularity":Popularity,"Vote_count":Vote_count,"Original_language":Original_language,"Vote_average":Vote_average}

    df=pd.DataFrame(d)
    
    final=final.append(df,ignore_index=True)
```

```python
final
```

```python
## Top rated title of the movie baased on the Popularity:
final[["Title","Popularity"]].groupby("Popularity").max().sort_values(by="Popularity",ascending=False)
```

```python
## Top rated title of the movie baased on the Vote_count:
final[["Title","Vote_count"]].groupby("Vote_count").max().sort_values(by="Vote_count",ascending=False)
```

```python
## Top rated title of the movie baased on the Vote_average:
final[["Title","Vote_average"]].groupby("Vote_average").max().sort_values(by="Vote_average",ascending=False)
```

```python
final["Release_date"]=pd.to_datetime(final["Release_date"])
```

```python
final["year"]=final["Release_date"].dt.year
```

```python
## No of movie release yearly :
final[["Title","year"]].groupby("year").count().sort_values(by="Title",ascending=False)
```
